```python
import sncosmo
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = sncosmo.Model(source='salt3')
```

```python
model.bandmag('desr', 'ab', [-10., 0., +10.])
```

```python
model.bandflux('desr', [-10., 0., +10.], zp=25., zpsys='ab')
```

```python
lamarr = np.arange(model.minwave(), model.maxwave()+1, 1)
```

```python
plt.plot(lamarr, model.flux(time=0, wave=lamarr))
plt.ylabel('Flux [ergs/s/cm^2/Angstrom]')
plt.xlabel('Wavelength [Angstrom]')
```
