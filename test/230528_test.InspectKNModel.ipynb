```python
import numpy as np
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
from astropy.time import Time
```

# Load the pickle

```python
import pickle
with open('../model/gw170817-like.interp.pkl', 'rb') as f:
	interp = pickle.load(f)
```

# Model parameters

```python
mdarr = np.array([0.001, 0.003, 0.01, 0.03, 0.1])
vdarr = np.array([0.05, 0.15, 0.3])
mwarr = np.array([0.001, 0.003, 0.01, 0.03, 0.1])
vwarr = np.array([0.05, 0.15, 0.3])
#	Viewing angle
angarr = np.linspace(0, 180, 54)

#	Phase
phasearr = np.array([0.125, 0.25, 0.5, 1.0])

#	Redshift
zarr = np.array([1e-4, 1e-3, 1e-2, 1e-1, 1e0])

mdmin = np.min(mdarr)
mdmax = np.max(mdarr)
vdmin = np.min(vdarr)
vdmax = np.max(vdarr)
mwmin = np.min(mwarr)
mwmax = np.max(mwarr)
vwmin = np.min(vwarr)
vwmax = np.max(vwarr)
angmin = np.min(angarr)
angmax = np.max(angarr)
phasemin = np.min(phasearr)
phasemax = np.max(phasearr)
zmin = np.min(zarr)
zmax = np.max(zarr)

print(f"md : {np.mean(mdarr):.3} ({mdmin:.3} - {mdmax:.3})")
print(f"vd : {np.mean(vdarr):.3} ({vdmin:.3} - {vdmax:.3})")
print(f"mw : {np.mean(mwarr):.3} ({mwmin:.3} - {mwmax:.3})")
print(f"vw : {np.mean(vwarr):.3} ({vwmin:.3} - {vwmax:.3})")
print(f"ang: {np.mean(angarr):.3} ({angmin:.3} - {angmax:.3})")
print(f"phase: {np.mean(phasearr):.3} ({phasemin:.3} - {phasemax:.3})")
```

- Wavelength array

```python
# lamarr = np.arange(1003, 127695.+1, 1)
lamarr = np.arange(2000, 11000.+1, 1)
isotropic_factor = 54.
```

# Comparison

## Dyn. Ejecta

- Mass Variation

```python
vd = vdarr[1]
mw = mwarr[1]
vw = vwarr[1]
ang = 45
phase = 0.5
```

```python
for md in mdarr:
	#	Input
	point = (
		md,
		vd,
		mw,
		vw,
		ang,
		phase,
		lamarr,
	)

	iflux = interp(point)*isotropic_factor
	plt.plot(lamarr, iflux, alpha=0.5, label=md)
plt.xlim([3000, 10000])
plt.xlabel(r'Wavelength [$\rm \AA$]')
plt.ylabel("Flux")
plt.legend()
```

- Velocity Variation

```python
md = mdarr[1]
# vd = vdarr[1]
mw = mwarr[1]
vw = vwarr[1]
ang = 45
phase = 0.5
for vd in vdarr:
	#	Input
	point = (
		md,
		vd,
		mw,
		vw,
		ang,
		phase,
		lamarr,
	)

	iflux = interp(point)*isotropic_factor
	plt.plot(lamarr, iflux, alpha=0.5, label=vd)
plt.xlim([3000, 10000])
plt.xlabel(r'Wavelength [$\rm \AA$]')
plt.ylabel("Flux")
plt.legend()
```
